# Stochastic Gradient Descent (SGD)

- Stochastic Gradient Descent (SGD) is a simple yet very efficient approach to discriminative learning of linear classifiers under convex loss functions such as (linear) Support Vector Machines and Logistic Regression. Even though SGD has been around in the machine learning community for a long time, it has received a considerable amount of attention just recently in the context of large-scale learning.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample dataset
X = np.array([1, 2, 3])  # Input features
y = np.array([2, 4, 6])  # True outputs
```

```python
plt.scatter(X, y, color='black', s=100)
plt.show()
```

$$\hat{y} = wx + b$$

- Mean Squared Error (MSE) Loss Function:
$$L = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

where:
- $L$ is the loss (error) function.
- $y_i$ is the actual value of the output.
- $\hat{y}$ is the predicted value: $\hat{y} = w x + b$.
- n is the number of training samples.

---

# Step-by-Step Derivation of $dw$ and $db$

Since Stochastic Gradient Descent (SGD) updates the model parameters for each individual data point, we calculate the gradient of the **loss function** with respect to each parameter (i.e., $w$ and $b$).

## 1. Compute Partial Derivative of the Loss Function w.r.t $w$
$$L = (y_i - \hat{y_i})^2$$

Using the **chain rule**:

$$\frac{dL}{dw} = 2(y_i - \hat{y_i}) \cdot \frac{d}{dw} (y_i - \hat{y_i})$$

Since $\hat{y_i} = w x_i + b$, we have:

$$\frac{d}{dw} (y_i - \hat{y_i}) = \frac{d}{dw} (y_i - (w x_i + b)) = -x_i$$

Thus, the gradient of the loss function with respect to $w$ is:

$$dw = -2 x_i (y_i - \hat{y_i})$$

---

## 2. Compute Partial Derivative of the Loss Function w.r.t $b$
Similarly, applying the chain rule:

$$\frac{dL}{db} = 2(y_i - \hat{y_i}) \cdot \frac{d}{db} (y_i - \hat{y_i})$$

Since:
$$\frac{d}{db} (y_i - \hat{y_i}) = \frac{d}{db} (y_i - (w x_i + b)) = -1$$

Thus, the gradient of the loss function with respect to $b$ is:
$$db = -2 (y_i - \hat{y_i})$$

```python
# Hyperparameters
learning_rate = 0.1
epochs = 100
```

```python
# Initialize parameters
w, b = 0, 0
```

$$\hat{y_i} = w x_i + b$$
$$dw = -2 x_i (y_i - \hat{y_i})$$
$$db = -2 (y_i - \hat{y_i})$$
$$w = w - \alpha \cdot dw$$
$$b = b - \alpha \cdot db$$

```python
# Stochastic Gradient Descent
for epoch in range(epochs):
    for i in range(len(X)):  # Iterate over each training sample
        xi, yi = X[i], y[i]
        
        # Prediction
        y_pred = w * xi + b
        
        # Compute gradients
        dw = -2 * xi * (yi - y_pred)
        db = -2 * (yi - y_pred)
        
        # Update weights
        w -= learning_rate * dw
        b -= learning_rate * db

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}: w = {w:.4f}, b = {b:.4f}")
```

```python
print("\nFinal Model: y =", round(w, 4), "* x +", round(b, 4))
```

```python
plt.scatter(X, y, color='black', s=100)
plt.plot(X, w*X + b, color='red', linewidth=2)
plt.show()
```
